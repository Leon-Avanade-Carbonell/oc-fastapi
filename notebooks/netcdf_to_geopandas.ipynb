```python
from pathlib import Path
import xarray as xr

# ── Set path to your NetCDF file ──────────────────────────────────────────
NC_FILE = Path("../local_only/1989.monthly_rain.nc")
```

```python
ds = xr.open_dataset(NC_FILE, engine="netcdf4")

print(f"File: {NC_FILE.resolve()}")
print(f"\nDimensions: {dict(ds.dims)}")

print("\nVariables:")
for name, var in ds.data_vars.items():
    units     = var.attrs.get("units", "—")
    long_name = var.attrs.get("long_name", "—")
    print(f"  {name:<24} shape={str(var.shape):<20} dtype={str(var.dtype):<10} units={units!r}  long_name={long_name!r}")

ds.close()
```
