# Database Access

Simple notebook for accessing the PostgreSQL database (oc-database).

```python
import psycopg2
import psycopg2.extras
import pandas as pd

# Database connection parameters
conn_params = {
    'host': 'localhost',
    'user': 'leon',
    'password': 'leon',
    'database': 'oc-database',
    'port': 5432
}

# Connect to database
conn = psycopg2.connect(**conn_params)
print(f"Connected to {conn_params['database']} as {conn_params['user']}")
```

## Query Tables

```python
# Get list of tables in the database
with conn.cursor() as cur:
    cur.execute("""
        SELECT tablename FROM pg_tables
        WHERE schemaname = 'public'
    """)
    tables = cur.fetchall()
    
print("Tables in oc-database:")
for table in tables:
    print(f"  - {table[0]}")
```
